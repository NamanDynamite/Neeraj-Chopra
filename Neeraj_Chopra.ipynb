```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df['Throw'] = df['Throw'].str.replace('m', '').astype(float)
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Throw'], bins=10, kde=True)
plt.title('Distribution of Throw Distances')
plt.xlabel('Throw Distance (m)')
plt.ylabel('Frequency')
mean_throw = df['Throw'].mean()
median_throw = df['Throw'].median()
plt.axvline(mean_throw, color='r', linestyle='--',
            label=f'Mean: {mean_throw:.2f} m')
plt.axvline(median_throw, color='g', linestyle='--',
            label=f'Median: {median_throw:.2f} m')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
df_sorted = df.sort_values(by='Date')
df_sorted.set_index('Date', inplace=True)
df_sorted
```

```python
df_sorted = df_sorted.drop('Competition', axis = 1)
df_sorted
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df_sorted.index, df_sorted['Throw'], marker='o',
linestyle='-', label='Thows')
plt.xticks(rotation=45)
for x, y in zip(df_sorted.index, df_sorted['Throw']):
  plt.annotate(f'{y}', (x, y), textcoords="offset points",
xytext=(0, 5), ha='center')
plt.title('Line Plot with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from datetime import datetime
df_sorted.index = pd.to_datetime(df_sorted.index)
df_sorted['Date_ordinal'] = df_sorted.index.map(datetime.toordinal)
X = df_sorted['Date_ordinal'].values.reshape(-1, 1)
y = df_sorted['Throw']
model = LinearRegression()
model.fit(X, y)
LinearRegression()
df_sorted['Predicted_Throw'] = model.predict(X)
plt.figure(figsize=(10, 6))
plt.plot(df_sorted.index, df_sorted['Throw'], label='Actual Throw',
marker='o')
plt.plot(df_sorted.index, df_sorted['Predicted_Throw'],
label='Predicted Throw', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Throw')
plt.title('Linear Regression on Throw Data Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
